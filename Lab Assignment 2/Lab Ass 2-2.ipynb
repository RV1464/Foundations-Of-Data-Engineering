```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Random heights
np.random.seed(42)
heights = np.random.randint(150, 191, 25)
df = pd.DataFrame({'Height': heights})
```

```python
# PMF and CDF
pmf_heights = df['Height'].value_counts(normalize=True).sort_index()
cdf_heights = pmf_heights.cumsum()
```

```python
# Plotting
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(pmf_heights.index, pmf_heights.values, width=1.0, edgecolor='black')
plt.title("PMF - Student Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Probability")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(cdf_heights.index, cdf_heights.values, where='mid')
plt.title("CDF - Student Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Cumulative Probability")
plt.grid(True)
```
