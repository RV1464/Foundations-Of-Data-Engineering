```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb
```

```python
# Binomial parameters
n = 10
p = 0.5
x_heads = np.arange(0, n + 1)

# PMF and CDF
pmf_heads = [comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in x_heads]
cdf_heads = np.cumsum(pmf_heads)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
markerline, stemlines, baseline = plt.stem(x_heads, pmf_heads)
plt.title("PMF - Number of Heads")
plt.xlabel("Heads")
plt.ylabel("Probability")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x_heads, cdf_heads, where='mid')
plt.title("CDF - Number of Heads")
plt.xlabel("Heads")
plt.ylabel("Cumulative Probability")
plt.grid(True)
```
