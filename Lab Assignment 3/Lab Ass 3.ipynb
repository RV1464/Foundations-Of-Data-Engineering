```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, expon
```

```python
x_norm = np.linspace(-4, 4, 500)
x_unif = np.linspace(0, 1, 500)
x_expo = np.linspace(0, 10, 500)
p_vals = np.linspace(0.001, 0.999, 1000)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

axs[0].plot(x_norm, norm.pdf(x_norm), label='PDF')
axs[0].plot(x_norm, norm.cdf(x_norm), label='CDF')
axs[0].plot(norm.ppf(p_vals), p_vals, label='PPF')
axs[0].set_title("Normal Distribution")
axs[0].legend()
axs[0].grid(True)

axs[1].plot(x_unif, uniform.pdf(x_unif), label='PDF')
axs[1].plot(x_unif, uniform.cdf(x_unif), label='CDF')
axs[1].plot(uniform.ppf(p_vals), p_vals, label='PPF')
axs[1].set_title("Uniform Distribution")
axs[1].legend()
axs[1].grid(True)

axs[2].plot(x_expo, expon.pdf(x_expo), label='PDF')
axs[2].plot(x_expo, expon.cdf(x_expo), label='CDF')
axs[2].plot(expon.ppf(p_vals), p_vals, label='PPF')
axs[2].set_title("Exponential Distribution")
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
