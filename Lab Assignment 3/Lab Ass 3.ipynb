```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, expon
```

```python
x_norm = np.linspace(-4, 4, 500)
x_unif = np.linspace(0, 1, 500)
x_expo = np.linspace(0, 10, 500)
p_vals = np.linspace(0.001, 0.999, 1000)
```

```python
fig, axs = plt.subplots(3, 3, figsize=(10, 12))

axs[0,0].plot(x_norm, norm.pdf(x_norm), label='PDF',color='blue')
axs[0,1].plot(x_norm, norm.cdf(x_norm), label='CDF',color='green')
axs[0,2].plot(norm.ppf(p_vals), p_vals, label='PPF',color='red')
axs[0,0].set_title("Normal PDF")
axs[0,1].set_title("Normal CDF")
axs[0,2].set_title("Normal PPF")

axs[1,0].plot(x_unif, uniform.pdf(x_unif), label='PDF',color='blue')
axs[1,1].plot(x_unif, uniform.cdf(x_unif), label='CDF',color='green')
axs[1,2].plot(uniform.ppf(p_vals), p_vals, label='PPF',color='red')
axs[1,0].set_title("Uniform PDF")
axs[1,1].set_title("Uniform CDF")
axs[1,2].set_title("Uniform PPF")

axs[2,0].plot(x_expo, expon.pdf(x_expo), label='PDF',color='blue')
axs[2,1].plot(x_expo, expon.cdf(x_expo), label='CDF',color='green')
axs[2,2].plot(expon.ppf(p_vals), p_vals, label='PPF',color='red')
axs[2,0].set_title("Exponential PDF")
axs[2,1].set_title("Exponential CDF")
axs[2,2].set_title("Exponential PPF")

for row in axs:
    for ax in row:
        ax.legend()
        ax.grid(True)
        
plt.tight_layout()
plt.show()
```
