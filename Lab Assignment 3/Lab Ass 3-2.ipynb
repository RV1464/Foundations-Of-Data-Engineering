```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_samples = 10000
```

```python
# Case i: X1 and X2 ~ Uniform(0, 1)
x1_i = np.random.uniform(0, 1, n_samples)
x2_i = np.random.uniform(0, 1, n_samples)
z_i = x1_i + x2_i
```

```python
# Case ii: X1 and X2 ~ Normal(0, 1)
x1_ii = np.random.normal(0, 1, n_samples)
x2_ii = np.random.normal(0, 1, n_samples)
z_ii = x1_ii + x2_ii
```

```python
# Case iii: X1 ~ Uniform(0, 1), X2 ~ Uniform(0, 2)
x1_iii = np.random.uniform(0, 1, n_samples)
x2_iii = np.random.uniform(0, 2, n_samples)
z_iii = x1_iii + x2_iii
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

axs[0].hist(z_i, bins=50, density=True, color='skyblue', edgecolor='black')
axs[0].set_title("Case i: Z = U(0,1) + U(0,1)")
axs[0].set_xlabel("Z")
axs[0].set_ylabel("Density")

axs[1].hist(z_ii, bins=50, density=True, color='purple', edgecolor='black')
axs[1].set_title("Case ii: Z = N(0,1) + N(0,1)")
axs[1].set_xlabel("Z")

axs[2].hist(z_iii, bins=50, density=True, color='lightgreen', edgecolor='black')
axs[2].set_title("Case iii: Z = U(0,1) + U(0,2)")
axs[2].set_xlabel("Z")

plt.tight_layout()
plt.show()
```
